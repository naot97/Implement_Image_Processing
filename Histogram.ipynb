# Histogram


```
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
```

```
im = data.camera()

#plot image
plt.imshow(im,cmap='gray', vmin=0, vmax=255)
plt.title("Image")
plt.show()

#get count
count = np.zeros((256,))
im_flat = im.flatten()
for i in im_flat:
  count[i] = count[i] + 1

#plot histogram
plt.plot(range(256), count)
plt.title("Histogram")
plt.show()
```
